# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
```

# Let's Generate quadractic data

```python
def noisy_quadratic(x : np.array, noise : float = 0.5):
    return x**2 + np.random.uniform(-1*noise,noise,size=x.shape)

x = np.linspace(-5,5,100)
y = noisy_quadratic(x, noise = 1.2)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y)
```

## Save This Dataset

```python
df = pd.DataFrame(dict(x=x,y=y))
df.to_csv(os.path.join('data','noisy_quadratic.csv'), index=False)
```

# Let's generate sinusoidal dataset

```python
def noisy_sine(x : np.array, noise : float = 0.5, random_state : int = 42):
    np.random.seed(random_state)
    return np.sin(x,) + np.random.uniform(-1*noise,noise,size=x.shape)

x = np.linspace(-10,10,300)
y = noisy_sine(x, noise = 0.3)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y)
```

```python
df = pd.DataFrame(dict(x=x,y=y))
df.to_csv(os.path.join('data','noisy_sine.csv'), index=False)
```

# Let's generate circular dataset for classification

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of points
num_points = 100

# Circle parameters
radius_inner = 1
radius_outer = 1.2

# Inner circle points (points inside the circle)
theta_inner = np.linspace(0, 2 * np.pi, num_points)
x_inner = radius_inner * np.cos(theta_inner) + np.random.normal(0, 0.1, num_points)
y_inner = radius_inner * np.sin(theta_inner) + np.random.normal(0, 0.1, num_points)

# Outer circle points (points on the circumference)
theta_outer = np.linspace(0, 2 * np.pi, num_points)
x_outer = radius_outer * np.cos(theta_outer)
y_outer = radius_outer * np.sin(theta_outer)
```

```python
# Plotting the points
plt.figure(figsize=(8, 8))
plt.scatter(x_inner, y_inner, alpha=0.6, label='Inside Circle (Class 1)')
plt.scatter(x_outer, y_outer, alpha=0.6, label='On Circle (Class 2)')
plt.title('2D Space: Circle Surrounded by Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show();
```

## Save this dataset

```python
x1 = np.hstack([x_inner, x_outer])
x2 = np.hstack([y_inner, y_outer])
target = np.array([0]*num_points + [1]*num_points)

df = pd.DataFrame(dict(x1=x1,x2=x2,target=target))
df.to_csv(os.path.join('data','noisy_circular.csv'), index=False)
```
