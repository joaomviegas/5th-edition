```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

# Part 1: Basic Exploration

```python
# Display the first few rows of the dataset
print("First few rows of the Iris dataset:")
print(iris.head())

# Summary statistics
print("\nSummary Statistics:")
print(iris.describe())
```

# Histograms of each feature

```python
# Histograms of each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris[feature], bins=20, kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

# Box plots to visualize quartiles and outliers

```python
# Box plots to visualize quartiles and outliers
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=feature, data=iris)
    plt.title(f'Box Plot of {feature} by Species')
plt.tight_layout()
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='species', y=feature, data=iris, inner='quart', palette='Set2')
    plt.title(f'Violin Plot of {feature} by Species')
plt.tight_layout()
plt.show()
```

# Correlation Matrix

```python
correlation_matrix = iris.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Pairplot to visualize pairwise relationships

```python
sns.pairplot(iris, hue='species', height=2)
plt.title('Pairplot of Iris Dataset')
plt.show()
```
